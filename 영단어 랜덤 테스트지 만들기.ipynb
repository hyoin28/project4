```python
import pandas as pd

#영단어 csv파일을 데이터프레임으로 불러오기
day1 = pd.read_csv('day1.csv')
day2 = pd.read_csv('day2.csv')

# display(day1.head())
# display(day2.head())

#pandas 라이브러리의 sample() 함수로 데이터프레임의 행을 랜덤으로 셔플하기
day1_shuffle = day1.sample(frac=1).reset_index(drop=True)
day2_shuffle = day2.sample(frac=1).reset_index(drop=True)

#답지 csv파일 저장하기
day1_shuffle.to_csv('day1_답지.csv', index=False, encoding="utf-8-sig")
day2_shuffle.to_csv('day2_답지.csv', index=False, encoding="utf-8-sig")

#테스지 만들기 - 속성 '뜻' 의 값들을 없애기
day1_shuffle['뜻'] = ''
day2_shuffle['뜻'] = ''

display(day1_shuffle.head()) #뜻이 지워졌는지 확인

#테스트지 csv파일 저장하기
day1_shuffle.to_csv('day1_test.csv', index=False, encoding="utf-8-sig")
day2_shuffle.to_csv('day2_test.csv', index=False, encoding="utf-8-sig")
```
